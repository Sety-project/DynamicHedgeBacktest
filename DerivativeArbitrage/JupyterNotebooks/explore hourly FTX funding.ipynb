```python
import pandas as pd
all_fundings = pd.read_parquet('C:/Users/Tibor/Sandbox/allfundings.parquet')
all_fundings
```

```python
import numpy as np
def moving_average(x, w):
    return np.concatenate(((w - 1)*[np.nan], np.convolve(x, np.ones(w), 'valid') / w))
%matplotlib notebook
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = 9, 6
funding = all_fundings['BTC-PERP/rate/funding'].dropna()
smooth = moving_average(funding, 24*7)
look_back = 24*365
plt.scatter(funding.index[-look_back:], funding[-look_back:], s = 1)
plt.scatter(funding.index[-look_back:], smooth[-look_back:], s = 1)
```

```python
fundings = all_fundings[-90*24:]
fundings['hour'] = fundings.index.hour
fundings = fundings.groupby(['hour']).median()
%matplotlib notebook
import matplotlib.pyplot as plt
plt.plot(fundings.mean(axis=1))
plt.xticks(range(24))
plt.axvline(x=10, color='r')
plt.axvline(x=22, color='g')
```

```python
all_fundings[-90*24:]
average_funding = all_fundings.mean(axis=1)
%matplotlib notebook
import matplotlib.pyplot as plt
plt.scatter(average_funding.index, average_funding, s=1)
import numpy as np
def moving_average(x, w):
    return np.concatenate(((w - 1)*[np.nan], np.convolve(x, np.ones(w), 'valid') / w))
smooth = moving_average(average_funding, 24*7)
plt.scatter(average_funding.index, smooth, s=1)
```

```python
fundings = all_fundings[-90*24:]
fundings['day_of_week'] = fundings.index.day_of_week
fundings = fundings.groupby(['day_of_week']).median()
%matplotlib notebook
import matplotlib.pyplot as plt
plt.plot(fundings.mean(axis=1))
```

```python
fundings = all_fundings[-90*24:]
fundings['moment'] = (100*fundings.index.day_of_week) + fundings.index.hour
fundings = fundings.groupby(['moment']).median()
%matplotlib notebook
import matplotlib.pyplot as plt
plt.plot(fundings.mean(axis=1))
```
