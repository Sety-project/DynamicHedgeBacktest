```python
import pandas as pd
spot = pd.read_parquet('C:/Users/Tibor/Data/20211205_58982_tob.parquet')
future = pd.read_parquet('C:/Users/Tibor/Data/20211205_75072_tob.parquet')
```

```python
spot['spot_bid'] = spot['bid_price']
spot['spot_ask'] = spot['ask_price']
spot = spot.set_index('exchange_timestamp_nanos')
spot = spot[['spot_bid', 'spot_ask']]
spot
```

```python
future['future_bid'] = future['bid_price']
future['future_ask'] = future['ask_price']
future = future.set_index('exchange_timestamp_nanos')
future = future[['future_bid', 'future_ask']]
future
```

```python
data = pd.merge(left=spot, right=future, how='outer', right_index=True, left_index=True)
data = data.sort_index()
data = data.ffill()
data = data.dropna(how='any')
data
```

```python
data['spot_mid'] = (data['spot_bid'] + data['spot_ask'])/2
data['future_mid'] = (data['future_bid'] + data['future_ask'])/2
data
```

```python
start = 1638718922000000000
end = 1638742654000000000
data_period = data[(data.index >= start) & (data.index <= end)]
data_period
```
