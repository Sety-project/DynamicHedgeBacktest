```python
from puffin import market_data as md
import datetime
import puffin.tools as tools
```

```python
symbol = 'ADAUSDT'
date_string = "20210901"
instrument_id = tools.get_id_from_symbol(symbol)
```

```python
data = md.MarketData(symbol)
date = datetime.date(int(date_string[0:4]),int(date_string[4:6]),int(date_string[6:8]))
data.load_top_of_book_data_from_parquet(date)
data.load_trade_data_from_parquet(date)
```

```python
data.generate_formatted_top_of_book_data()
```

```python
data.generate_formatted_trades_data()
```

```python
tob_formatted = data.top_of_book_formatted
tob_formatted
```

```python
trades_formatted = data.trades_formatted
trades_formatted
```

```python
tob_formatted.to_csv(f'../data/inputs/{date_string}_{instrument_id}_{symbol}_tobs.csv', index=False)
trades_formatted.to_csv(f'../data/inputs/{date_string}_{instrument_id}_{symbol}_trades.csv', index=False)
```
